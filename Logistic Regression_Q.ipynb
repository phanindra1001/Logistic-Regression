```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("P:/python/bank-full.csv", sep = ';')
data.tail()
```

```python
data.info()
```

### EDA

```python
Month = {'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}
data.month = data.month.map(Month)
```

```python
data.default = data.default.map(dict(yes=1, no=0))
data.housing = data.housing.map(dict(yes=1, no=0))
data.loan = data.loan.map(dict(yes=1, no=0))
data.y = data.y.map(dict(yes=1, no=0))
```

```python
data.head()
```

```python
data = pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'])
```

```python
data.head()
```

```python
pd.options.display.max_columns = None
data
```

```python
data.info()
```

```python
x = pd.concat([data.iloc[:,:11],data.iloc[:,12:]],axis = 1)
y = data.iloc[:,11]
x
```

### Model building

```python
classifier = LogisticRegression()
classifier.fit(x,y)
```

### Model prediction

```python
preds = classifier.predict(x)
preds
```

```python
y_preds= pd.DataFrame({'actual': y,'predicted_prob': preds})
y_preds
```

### model evaluation

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,preds)
print (confusion_matrix)
```
